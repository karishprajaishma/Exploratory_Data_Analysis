```python
import pandas as pd
```

```python
dataset=pd.read_csv("claimants.csv")
dataset.head()
```

```python
dataset.isna().sum()
```

```python
dataset['CLMSEX']=dataset['CLMSEX'].fillna(dataset['CLMSEX'].mode,inplace=False)
dataset['CLMINSUR']=dataset['CLMINSUR'].fillna(dataset['CLMINSUR'].mode,inplace=False)
dataset['SEATBELT']=dataset['SEATBELT'].fillna(dataset['SEATBELT'].mode,inplace=False)
dataset['CLMAGE']=dataset['CLMAGE'].fillna(dataset['CLMAGE'].mean(),inplace=False)
```

```python
dataset.isna().sum()
```

```python
dataset.describe()
```

```python
#z-score = x-mean/std.deviation
mu=dataset['CLMAGE'].mean()
sd=dataset['CLMAGE'].std()
print("Mean: ",mu)
print("Standard Deviation: ",sd)
```

```python
z_score=(dataset['CLMAGE']-mu)/sd
dataset['Z_SCORE']=z_score
dataset.head()
```

```python
outliers=dataset[dataset['Z_SCORE'].abs() > 3]
print(outliers)
```

```python
dataset=dataset.drop(columns=['Z_SCORE'])
dataset.head()
```
